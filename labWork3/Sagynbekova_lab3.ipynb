```python
# Коэффициенты квадратичной функции
mpl.rcParams['font.family']='fantasy'
a = -10
b = 2
c = 3
# Создаем массив значений x
x = np.linspace(-20,20,100)  

# Вычисляем значения y для каждого x
y = a * x**2 + b * x + c

# Создаем график

plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')  # Построение графика
plt.xlabel('LINE X')  
plt.ylabel('LINE Y') 
plt.title('График')  
plt.show()
```

```python
# Задаем параметры гармонической функции
amplitude = 1.0  # Амплитуда
frequency = 2.0  # Частота
phase = np.pi / 4  # Фаза
x_min = 0
x_max = 2 * np.pi  # Диапазон значений x 

# Создаем массив значений x
x = np.linspace(x_min, x_max, 400)

# Вычисляем значения y для каждого x
y = amplitude * np.sin(frequency * x + phase)

# Создаем график
plt.plot(x, y, label=f'y = {amplitude} * sin({frequency}x + {phase:.2f})')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('График гармонической функции')
plt.show()
```

Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
# Коэффициенты квадратичной функции
a = -10
b = 2
c = 3

x = np.linspace(-20, 20, 100)

y = a * x**2 + b * x + c

amplitude = 1.0
frequency = 2.0
phase = np.pi / 4
x_har = np.linspace(-20, 20, 100)

y_har = amplitude * np.sin(frequency * x_har + phase)

# Первый график (квадратичная функция)
plt.subplot(1, 2, 1)  # 1 строка, 2 столбца, первый график
plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('График квадратичной функции')
plt.grid(color = 'r')
plt.legend()

# Второй график (гармоническая функция)
plt.subplot(1, 2, 2)  # 1 строка, 2 столбца, второй график
plt.plot(x_har, y_har, label=f'y = {amplitude} * sin({frequency}x + {phase:.2f})')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('График гармонической функции')
plt.grid(True)
plt.legend()

plt.show()
```

Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи

```python
a = -10
b = 2
c = 3

x = np.linspace(-5, 5, 100)

y = a * x**2 + b * x + c

amplitude = 15.0
frequency = 10.0
phase = np.pi / 4

y_har = amplitude * np.sin(frequency * x + phase)

# Создаем график
plt.figure()

# График квадратичной функции
plt.plot(x, y, label=f'Квадратичная: y = {a}x^2 + {b}x + {c}')

# График гармонической функции
plt.plot(x, y_har, label=f'Гармоническая: y = {amplitude} * sin({frequency}x + {phase:.2f})')

# Добавляем подписи
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Графики квадратичной и гармонической функций')

plt.legend()
plt.grid(True)
# Отображаем графики
plt.show()
```

Построить график произвольной каркасной поверхности.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Создаем фигуру и 3D-подграфик
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Определите диапазон значений x и y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Создайте сетку значений x и y
X, Y = np.meshgrid(x, y)

# Определите функцию z в зависимости от x и y (произвольная поверхность)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Построение каркасной поверхности
ax.plot_wireframe(X, Y, Z, color='b', linewidth=0.5)

# Добавьте подписи к осям
ax.set_xlabel('X-ось')
ax.set_ylabel('Y-ось')
ax.set_zlabel('Z-ось')

# Заголовок графика
plt.title('Произвольная каркасная поверхность')

# Отображение графика
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Создаем фигуру и 3D-подграфик
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Генерируем данные для каркасной поверхности
u = np.linspace(0, 5 * np.pi, 200)
v = np.linspace(0, np.pi, 200)

x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

# Построение каркасной поверхности
ax.plot_wireframe(x, y, z, color='r', linewidth=0.5)

# Добавляем подписи к осям
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Произвольная каркасная поверхность')
plt.show()
```
